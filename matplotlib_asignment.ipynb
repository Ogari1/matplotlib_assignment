```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

#Load the dataset into a Pandas DataFrame
df = pd.read_csv("jumta.csv")
print(df)
```

```python
#line plot showing the monthly revenue over the two-year period
plt.figure(figsize=(10,10))
plt.plot(df['Month'], df['Revenue'], color='red', linestyle='--', marker='s', linewidth=2, label='Revenue')
plt.title('Monthly Revenue 2020-2021')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(color='gray', linestyle='--')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
#bar plot showing the total number of orders per month
plt.figure(figsize=(10,10))
plt.bar(df['Month'], df['Orders'], color='pink', linestyle='--', label='Orders')
plt.title('Monthly Orders 2020-2021')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.grid(axis='y', color='gray', linestyle='--')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
#scatter plot showing the relationship between the number of orders and the average order value
plt.figure(figsize=(10,5))
plt.scatter(df['Orders'], df['Average_Order_Value'], color='purple', linestyle='--', label='Orders')
plt.title('Number of orders vs Average Order Value (2020-2021)')
plt.xlabel('Number of Orders')
plt.ylabel('Average Order Value')
plt.grid(axis='y', color='gray', linestyle='--')
plt.legend()
plt.show()
```

```python
# Create a subplot with all three visualizations
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10, 15))
# Line plot of monthly revenue

ax1.plot(df['Month'], df['Revenue'], color='red', linestyle='--', marker='s', linewidth=2, label='Revenue')
ax1.set_title('Monthly Revenue 2020-2021')
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue')
ax1.grid(color='gray', linestyle='--')
ax1.tick_params(axis='x', rotation=90)
ax1.legend()

# Bar plot of total number of orders per month

ax2.bar(df['Month'], df['Orders'], color='pink', linestyle='--', label='Orders')
ax2.set_title('Monthly Orders 2020-2021')
ax2.set_xlabel('Month')
ax2.set_ylabel('Number of Orders')
ax2.grid(axis='y', color='gray', linestyle='--')
ax2.tick_params(axis='x', rotation=90)
ax2.legend()

# Scatter plot of number of orders vs average order value

ax3.scatter(df['Orders'], df['Average_Order_Value'], color='purple', linestyle='--', label='Orders')
ax3.set_title('Number of orders vs Average Order Value (2020-2021)')
ax3.set_xlabel('Number of Orders')
ax3.set_ylabel('Average Order Value')
ax3.grid(axis='y', color='gray', linestyle='--')
ax3.legend()


plt.tight_layout()
plt.savefig('sales_data_analysis.png', dpi=300)
plt.show()
```
